# GeoPandas Tutorial

```python
import matplotlib.pyplot as plt
import geopandas
from cartopy import crs as ccrs
```

```python
path = geopandas.datasets.get_path('naturalearth_lowres')

df = geopandas.read_file(path)

# Add a column that'll be used later #
df['gdp_pp'] = df['gdp_md_est'] / df['pop_est']

```

```python
# Visualizing the map with GeoPandas #
df.plot()
```

# Plotting with CartoPy #

```python
# Defining the CartoPy CRS Object #
crs = ccrs.AzimuthalEquidistant()

# Can be converted into a proj4 string/dict compatible with geopandas #
crs_proj4 = crs.proj4_init 
df_ae = df.to_crs(crs_proj4)

# Here's what the plot looks like in geopandas #
df_ae.plot()
```

```python
# Now with data in CRS based on CartoPy we can easily plot it # ####Doesn't work #####
fig, ax = plt.subplots(subplot_kw = {'projection': crs})
ax.add_geometries(df_ae['geometry'], crs = crs)
```
